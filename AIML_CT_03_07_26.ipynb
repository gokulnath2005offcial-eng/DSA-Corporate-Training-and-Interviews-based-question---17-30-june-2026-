```python
# Complex Data Type in Python

# Creating Complex Numbers
num1 = 3 + 4j
num2 = 5 + 2j

# Display Complex Numbers
print("Complex Number 1:", num1)
print("Complex Number 2:", num2)

# Check Data Type
print("Data Type of num1:", type(num1))
print("Data Type of num2:", type(num2))

# Arithmetic Operations
print("\nArithmetic Operations")
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)

# Real and Imaginary Parts
print("\nReal and Imaginary Parts")
print("Real Part of num1:", num1.real)
print("Imaginary Part of num1:", num1.imag)

# Creating Complex Number using complex() Function
num3 = complex(10, 20)

print("\nComplex Number using complex() Function:", num3)
print("Real Part:", num3.real)
print("Imaginary Part:", num3.imag)

# Comparison
print("\nComparison")
print("num1 == num2:", num1 == num2)
print("num1 != num2:", num1 != num2)
```

Python List Methods:

```python
# Creating a list
fruits = ["Apple", "Banana", "Cherry"]

print("Original List:", fruits)

# append()
fruits.append("Mango")
print("append():", fruits)

# extend()
fruits.extend(["Orange", "Grapes"])
print("extend():", fruits)

# insert()
fruits.insert(1, "Pineapple")
print("insert():", fruits)

# remove()
fruits.remove("Banana")
print("remove():", fruits)

# pop()
item = fruits.pop()
print("pop():", item)
print(fruits)

# index()
print("index of Apple:", fruits.index("Apple"))

# count()
numbers = [10, 20, 10, 30, 10]
print("count of 10:", numbers.count(10))

# sort()
numbers.sort()
print("sort():", numbers)

# reverse()
numbers.reverse()
print("reverse():", numbers)

# copy()
new_numbers = numbers.copy()
print("copy():", new_numbers)

# clear()
new_numbers.clear()
print("clear():", new_numbers)
```

Tuples Methods:

```python
# Creating a Tuple
fruits = ("Apple", "Banana", "Cherry", "Mango", "Orange")

print("Original Tuple:", fruits)

# Accessing Elements
print("\nFirst Element:", fruits[0])
print("Last Element:", fruits[-1])

# Slicing
print("\nFirst 3 Elements:", fruits[:3])
print("Last 2 Elements:", fruits[-2:])

# Length of Tuple
print("\nLength:", len(fruits))

# Membership Operators
print("\nIs Apple present?", "Apple" in fruits)
print("Is Grapes present?", "Grapes" in fruits)

# Concatenation
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
result = tuple1 + tuple2
print("\nConcatenation:", result)

# Repetition
repeat = tuple1 * 2
print("Repetition:", repeat)

# count() Method
numbers = (10, 20, 30, 10, 40, 10)
print("\nCount of 10:", numbers.count(10))

# index() Method
print("Index of 30:", numbers.index(30))

# Built-in Functions
print("\nMaximum:", max(numbers))
print("Minimum:", min(numbers))
print("Sum:", sum(numbers))

# Sorted Function
print("Sorted:", sorted(numbers))

# Iterating Through Tuple
print("\nElements in Tuple:")
for item in fruits:
    print(item)

# Nested Tuple
student = ("Gokul", 20, ("Python", "SQL", "AI"))
print("\nNested Tuple:", student)
print("Course 1:", student[2][0])
print("Course 2:", student[2][1])

# Packing
packed = 100, 200, 300
print("\nPacked Tuple:", packed)

# Unpacking
a, b, c = packed
print("Unpacked Values:", a, b, c)

print("\nProgram Completed Successfully!")
```

Dictionary Methods:

```python
# Creating a Dictionary
student = {
    "Name": "Gokul",
    "Age": 20,
    "Course": "Python",
    "Marks": 95
}

print("Original Dictionary:")
print(student)

# Accessing Values
print("\nName:", student["Name"])
print("Course:", student.get("Course"))

# keys()
print("\nKeys:")
print(student.keys())

# values()
print("\nValues:")
print(student.values())

# items()
print("\nItems:")
print(student.items())

# update()
student.update({"Age": 21, "City": "Chennai"})
print("\nAfter update():")
print(student)

# setdefault()
student.setdefault("College", "ABC College")
print("\nAfter setdefault():")
print(student)

# pop()
student.pop("Marks")
print("\nAfter pop('Marks'):")
print(student)

# popitem()
student.popitem()
print("\nAfter popitem():")
print(student)

# copy()
new_student = student.copy()
print("\nCopied Dictionary:")
print(new_student)

# Length
print("\nLength:", len(student))

# Membership
print("\nIs 'Name' present?", "Name" in student)

# Looping
print("\nDictionary Elements:")
for key, value in student.items():
    print(key, ":", value)

# clear()
student.clear()
print("\nAfter clear():")
print(student)
```
