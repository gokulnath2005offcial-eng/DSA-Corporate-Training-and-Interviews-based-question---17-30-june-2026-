Python List Methods:

```python
# Creating a list
fruits = ["Apple", "Banana", "Cherry"]

print("Original List:", fruits)

# append()
fruits.append("Mango")
print("append():", fruits)

# extend()
fruits.extend(["Orange", "Grapes"])
print("extend():", fruits)

# insert()
fruits.insert(1, "Pineapple")
print("insert():", fruits)

# remove()
fruits.remove("Banana")
print("remove():", fruits)

# pop()
item = fruits.pop()
print("pop():", item)
print(fruits)

# index()
print("index of Apple:", fruits.index("Apple"))

# count()
numbers = [10, 20, 10, 30, 10]
print("count of 10:", numbers.count(10))

# sort()
numbers.sort()
print("sort():", numbers)

# reverse()
numbers.reverse()
print("reverse():", numbers)

# copy()
new_numbers = numbers.copy()
print("copy():", new_numbers)

# clear()
new_numbers.clear()
print("clear():", new_numbers)
```

Tuples Methods:

```python
# Creating a Tuple
fruits = ("Apple", "Banana", "Cherry", "Mango", "Orange")

print("Original Tuple:", fruits)

# Accessing Elements
print("\nFirst Element:", fruits[0])
print("Last Element:", fruits[-1])

# Slicing
print("\nFirst 3 Elements:", fruits[:3])
print("Last 2 Elements:", fruits[-2:])

# Length of Tuple
print("\nLength:", len(fruits))

# Membership Operators
print("\nIs Apple present?", "Apple" in fruits)
print("Is Grapes present?", "Grapes" in fruits)

# Concatenation
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
result = tuple1 + tuple2
print("\nConcatenation:", result)

# Repetition
repeat = tuple1 * 2
print("Repetition:", repeat)

# count() Method
numbers = (10, 20, 30, 10, 40, 10)
print("\nCount of 10:", numbers.count(10))

# index() Method
print("Index of 30:", numbers.index(30))

# Built-in Functions
print("\nMaximum:", max(numbers))
print("Minimum:", min(numbers))
print("Sum:", sum(numbers))

# Sorted Function
print("Sorted:", sorted(numbers))

# Iterating Through Tuple
print("\nElements in Tuple:")
for item in fruits:
    print(item)

# Nested Tuple
student = ("Gokul", 20, ("Python", "SQL", "AI"))
print("\nNested Tuple:", student)
print("Course 1:", student[2][0])
print("Course 2:", student[2][1])

# Packing
packed = 100, 200, 300
print("\nPacked Tuple:", packed)

# Unpacking
a, b, c = packed
print("Unpacked Values:", a, b, c)

print("\nProgram Completed Successfully!")
```

Dictionary Methods:

```python
# Creating a Dictionary
student = {
    "Name": "Gokul",
    "Age": 20,
    "Course": "Python",
    "Marks": 95
}

print("Original Dictionary:")
print(student)

# Accessing Values
print("\nName:", student["Name"])
print("Course:", student.get("Course"))

# keys()
print("\nKeys:")
print(student.keys())

# values()
print("\nValues:")
print(student.values())

# items()
print("\nItems:")
print(student.items())

# update()
student.update({"Age": 21, "City": "Chennai"})
print("\nAfter update():")
print(student)

# setdefault()
student.setdefault("College", "ABC College")
print("\nAfter setdefault():")
print(student)

# pop()
student.pop("Marks")
print("\nAfter pop('Marks'):")
print(student)

# popitem()
student.popitem()
print("\nAfter popitem():")
print(student)

# copy()
new_student = student.copy()
print("\nCopied Dictionary:")
print(new_student)

# Length
print("\nLength:", len(student))

# Membership
print("\nIs 'Name' present?", "Name" in student)

# Looping
print("\nDictionary Elements:")
for key, value in student.items():
    print(key, ":", value)

# clear()
student.clear()
print("\nAfter clear():")
print(student)
```

Linear algebra :

```python
import numpy as np

# Matrix A
A = np.array([[1, 2],
              [3, 4]])

# Matrix B
B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

# Addition
print("\nMatrix Addition:")
print(A + B)

# Subtraction
print("\nMatrix Subtraction:")
print(A - B)

# Multiplication
print("\nMatrix Multiplication:")
print(np.dot(A, B))

# Transpose
print("\nTranspose of Matrix A:")
print(A.T)

# Determinant
print("\nDeterminant of Matrix A:")
print(np.linalg.det(A))

# Inverse
print("\nInverse of Matrix A:")
print(np.linalg.inv(A))

# Rank
print("\nRank of Matrix A:")
print(np.linalg.matrix_rank(A))

# Trace
print("\nTrace of Matrix A:")
print(np.trace(A))

# Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("\nEigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)

# Dot Product
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print("\nVector 1:")
print(v1)

print("\nVector 2:")
print(v2)

print("\nDot Product:")
print(np.dot(v1, v2))

# Cross Product
print("\nCross Product:")
print(np.cross(v1, v2))

# Matrix Norm
print("\nNorm of Matrix A:")
print(np.linalg.norm(A))

# Solve Linear Equation
C = np.array([[2, 1],
              [1, 3]])

D = np.array([8, 13])

solution = np.linalg.solve(C, D)

print("\nMatrix C:")
print(C)

print("\nMatrix D:")
print(D)

print("\nSolution of Linear Equation:")
print(solution)
```

Standard Deviation :

```python
import numpy as np

# Get the number of elements
n = int(input("Enter the number of elements: "))

# Create an empty list
data = []

# Read the elements
for i in range(n):
    value = float(input(f"Enter element {i+1}: "))
    data.append(value)

# Convert the list to a NumPy array
arr = np.array(data)

# Calculate statistics
mean = np.mean(arr)
variance = np.var(arr)
standard_deviation = np.std(arr)

# Display the results
print("\nArray:", arr)
print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", standard_deviation)
```

Pandas Row Selecting Code :

```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Gokul", "Naveen", "Sanjay", "Vishnu"],
    "Age": [20, 21, 22, 23],
    "Marks": [85, 90, 75, 95],
    "Department": ["IT", "CSE", "ECE", "MECH"]
}

df = pd.DataFrame(data)

# Display complete DataFrame
print("Original DataFrame:")
print(df)


# 1. Select first row using iloc
print("\nFirst Row:")
print(df.iloc[0])


# 2. Select multiple rows using iloc
print("\nFirst Two Rows:")
print(df.iloc[0:2])


# 3. Select last row using iloc
print("\nLast Row:")
print(df.iloc[-1])


# 4. Select row using loc
print("\nRow with Index 2:")
print(df.loc[2])


# 5. Select specific rows and columns
print("\nSelect Rows and Columns:")
print(df.loc[0:2, ["Name", "Marks"]])


# 6. Select rows using condition
print("\nStudents Marks Greater Than 80:")
result = df[df["Marks"] > 80]
print(result)


# 7. Select rows based on age
print("\nStudents Age Greater Than 21:")
result = df[df["Age"] > 21]
print(result)
```
